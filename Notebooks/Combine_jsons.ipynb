```python
files = glob.glob(r'C:\Users\Brands\Documents\Cardiff\Dissertation work\Data_analysis\Data\June_2022\/*')
```

```python
pprint(files)
```

```python
# read each line of the files, extract relevant data
final = []
rows = []
for name in files:
    with open(name, "r") as f:
        for line in f:
            try:
                data = json.loads(line)
            except:
                pass
                
            tweet_id = data["id_str"]

            is_reply = data["in_reply_to_screen_name"]

            is_retweet = data["retweeted"]

            user_name = data["user"]["screen_name"]

            hashtags = []
            for hashtag in data["entities"]["hashtags"]:
                hashtags.append(hashtag["text"])

            if "extended_tweet" in data:
                text = data["extended_tweet"]["full_text"].lower()
            elif data["text"]:
                text = data["text"].lower()

            created = data["created_at_src"]

            rows.append((tweet_id, hashtags, text, is_reply, is_retweet, user_name, created))
                
        final.extend(rows)
```

```python
df = pd.DataFrame(data=final, columns=["tweet_id", "hashtags", "tweet_text", "is_reply", "is_retweet", "user_name", "created"])
```
