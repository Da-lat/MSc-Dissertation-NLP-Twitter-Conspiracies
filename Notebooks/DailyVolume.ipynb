# Average volume for each month

```python
avg_tweets = [12507, 8015, 8975, 3380, 9730, 11717, 12919, 15630, 20522]
months = ["May 2019", "June 2019", "May 2020", "June 2020", "May 2021", "June 2021", "May 2022", "June 2022", "July 2022"]
percent = [0.42, 0.34, 0.13, 0.10, 0.22, 0.31, 0.23, 0.39, 0.52]

sns.barplot(x=months, y=percent)
plt.title("Percentage of tweets that were climate related for each months dataset from May 2019 to July 2022", fontsize=18)
plt.ylabel("Percentage of climate related tweets (%)", fontsize=18)
plt.xlabel("Month and year", fontsize=18)
plt.savefig("percent_tweets.png")
plt.show()
```

```python
avg_daily = sns.barplot(x=months, y=avg_tweets)
plt.title("Average daily volume of climate related tweets per month", fontsize=18)
plt.ylabel("Average daily tweets", fontsize=18)
plt.xlabel("Month and year", fontsize=18)
plt.savefig("Average_daily_tweets.png")
plt.show()
```
